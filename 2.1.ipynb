```python
import math
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
a = 9.0
b = 8.0
c = 8.0

r = np.linspace(-2, 2, 100)
w = ((a - 4) * (r ** 2) + (b - 5) * r + (c - 6))

z = (math.e**r) / (math.e ** r) + 1

plt.plot(r, w, 'c')
plt.plot(r, z, 'r')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
